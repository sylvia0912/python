## Word Cloud about trending games category - action,sports, racing, educational etc

### Application of Word count
Say the firm received 1000 feedbacks from different users. It would be very difficult for the stakeholders to read and make note of every feedback. Thus, Word Cloud would play a key role in getting top keywords among the feedback. This would help the firm determine if the feedback is positives or negatives and respective areas of improvement.

```python
import os

import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
df = pd.read_csv("/Users/sylvia/Desktop/datasets/android-games.csv")
df.head()
```

```python
df['category'].value_counts()
```

```python
import os
os.getcwd()
```

```python
df.isna().sum()
```

```python
text = " ".join(cat.split()[1] for cat in df.category)
```

```python
word_cloud = WordCloud(collocations = False, background_color = 'white').generate(text)
```

```python
plt.imshow(word_cloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
